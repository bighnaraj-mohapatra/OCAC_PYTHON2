```python
#Generating a random integer between 2 given numbers(inclusive)
random.randint(5,6)
```

```python
#Generating a random integer within a range , we use randrange()
random.randrange(2,20,2)
```

```python
#How to generate 5 random numbers
for i in range(5):
    print(random.randrange(5))
```

```python
# choice() returns a random object
numbers = [1,24,'Bighnaraj','Amrit',32,56]
print(random.choice(numbers))
```
